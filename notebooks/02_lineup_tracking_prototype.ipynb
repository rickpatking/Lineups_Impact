```python
subs = df[df['actionType'] == 'Substitution'].copy()

subs['prev_name'] = subs['description'].str.split(r': | FOR').str[2]
subs['prev_name'] = subs['prev_name'].str.lstrip()
subs['prev_id'] = subs['personId']
subs['next_name'] = subs['description'].str.split(r': | FOR').str[1]
subs['next_name'] = subs['next_name'].str.lstrip()
subs['next_id'] = None
subs
```

start by getting only the pacer subs' ids

```python
subs_pacers = subs[subs['teamTricode'] == 'IND'].copy()
sorted_prev = np.sort(subs_pacers['prev_name'].unique())
sorted_next = np.sort(subs_pacers['next_name'].unique())
print(sorted_prev)
print(sorted_next)
```

If the players in and out match, you don't need to get any ids from the team

```python
intersection = np.intersect1d(sorted_prev, sorted_next)
difference = np.setdiff1d(sorted_prev, sorted_next)
difference2 = np.setdiff1d(sorted_next, sorted_prev)
print(intersection)
print(difference)
print(difference2)
all_pacers_players = np.concatenate((intersection, difference, difference2), axis=0)
all_pacers_players
```

```python
player_id_dict = dict()
for player in all_pacers_players:
    try:
        playerdf = df[df['playerName'] == player]
        id = ((playerdf.iloc[0])['personId'])
        player_id_dict[player] = id
    except Exception as e: 
        team_roster = CommonTeamRoster(team_id=1610612754, season='2024-25')
        roster_df = team_roster.get_data_frames()[0]
        roster_df2 = (roster_df[roster_df['PLAYER'].str.contains(player)]).copy()
        player_id_dict[player] = roster_df2['PLAYER_ID'].iloc[0]
player_id_dict
```

was able to get every player id for the pacers

```python
for player, id in player_id_dict.items():
    subs.loc[subs['next_name'] == player, 'next_id'] = id
subs
```

```python
all_lineups = []
all_lineups.append(starting_5)
pacer_subs = subs.loc[subs['teamTricode'] == 'IND']
for index, sub in pacer_subs.iterrows():
    try:
        current_five = all_lineups[-1][:]
        print(current_five)
        print(index)
        current_five.remove(sub['prev_id'])
        current_five.append(sub['next_id'])
        all_lineups.append(current_five)
    except Exception as e:
        print(e)
    print()
all_lineups
```
